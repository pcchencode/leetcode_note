# Question 121. Best Time to Buy and Sell Stock
Say you have an array for which the ith element is the price of a given stock on day i.

If you were only permitted to complete at most one transaction (i.e., buy one and sell one share of the stock), design an algorithm to find the maximum profit.

Note that you cannot sell a stock before you buy one.

Example 1:

Input: [7,1,5,3,6,4]
Output: 5
Explanation: Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = 6-1 = 5.
             Not 7-1 = 6, as selling price needs to be larger than buying price.
Example 2:

Input: [7,6,4,3,1]
Output: 0
Explanation: In this case, no transaction is done, i.e. max profit = 0.

```python
import time
import random
```

## 暴力解

```python
def maxProfit(prices):
    profit_choice = []
    for i in range(0,len(prices)):
        for j in range(i+1,len(prices)):
            tmp = []
            tmp.append(prices[j]-prices[i])
            profit_choice.append(max(tmp))
    
    try:
        if max(profit_choice) < 0:
            return 0
        else:
            return max(profit_choice)
    except ValueError:
        return 0
```

```python
print(maxProfit([7,1,6,5,4]))
print(maxProfit([7,6,4,3,1]))
print(maxProfit([]))
```

## Dynamic Programming
* 個別選原始 prices 中每一個買入點，針對個別買入點都選擇一個 optimal selling price

* 將選擇的 optimal selling price 計算 optimal profit，塞入 profit_choice 當中

* profit_choice set 中取 max，即為所求

```python
def maxProfitDP(prices):
    n = len(prices)
    profit_choice = []
    for i in range(0, n-1): # 因為不可能選在最後一個 tick 買
        profit_choice.append(max(prices[i+1:n]) - prices[i])

    if len(profit_choice)==0 or max(profit_choice)<0:
        return 0
    else:
        return max(profit_choice)
```

```python
print(maxProfitDP([7,1,6,5,4]))
print(maxProfitDP([7,6,4,3,1]))
print(maxProfitDP([]))
```

## 來看看 Dynamic Programming 的威力吧

```python
# 造一個超級長的隨機 integer array
def get_price_arr(length):
    prices = []
    for i in range(length):
        prices.append(random.randint(1,100))

    return prices
```

```python
prices = get_price_arr(10000)
print(prices)
```

```python
t_start = time.time()
maxProfit(prices)
t_end = time.time()
print('It costs '+str(t_end - t_start)+' secs')
# 暴力解要耗時 15 秒
```

```python
t_start = time.time()
maxProfitDP(prices)
t_end = time.time()
print('It costs '+str(t_end - t_start)+' secs')
# 動態規劃近乎秒解
```
